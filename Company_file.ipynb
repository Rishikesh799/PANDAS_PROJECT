# 2].a)Add column lead time : lead time is difference between the ordered date and the received date in days

```python
x=data["Purchase Orders 1"]
```

```python
x
```

```python
x["lead_time"]=x["Received Date"]-x["Ordered Date"]
x["lead_time"]=x["lead_time"]/np.timedelta64(1,"D")  ##-->np.timedelta(1,"D") instead of D WE use W--WEEK,M--MONTH
x
```

# b)Calculate lead time average and lead time variance for each Product id-branch id-Supplier id combination.

```python
x.groupby(["Product ID","Branch ID","Supplier ID"])["lead_time"].agg([np.mean,np.var])
```

```python
n={"monday":0,"tuesday":1,"wednesday":2,"thursday":3,"friday":4,"saturday":5,"sunday":6}
```

```python
def chang(x):
    if len(x)==1:
        for j in x:
            return str(n[j.lower()])
    else:
        d=[]
        for i in x:
            d.append(n[i.lower()])
        return str(d)
```
